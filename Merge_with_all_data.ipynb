```python
import pandas as pd
import numpy as np
import geopandas as gpd
# from shapely.geometry import Polygon
import os
import datetime
import matplotlib.pyplot as plt
import contextily as cx
import seaborn as sns
```

```python
# 25GB数据读取
df_raw=pd.read_csv('/Users/zheng/Desktop/时空数据挖掘/Data/data/sample 10000.csv', usecols= [0,2,3,5, 28,29, 31,32, 33, 34 ,35], nrows=10000)
name_list=df_raw.columns.tolist()
dfs = []
chunk_size = 100000
for chunk in pd.read_csv('/Users/zheng/Desktop/时空数据挖掘/Data/data/000000_0_fb744a52-2d44-4e8d-8ff1-85aeda22a123_20231207_153237_00071_9h6es', chunksize=chunk_size,error_bad_lines=False,usecols= [0,2,3,5, 28,29, 31,32, 33, 34, 35 ],header= None,names=name_list):
    dfs.append(chunk)
df= pd.concat(dfs)
```

```python
# TPU 读取
gfs = []
for f in os.listdir('/Users/zheng/Desktop/时空数据挖掘/Data/tpu_data_geo/'):
    if 'shp' not in f:
        continue
    if '._' in f:
        continue
    ## GBK for encode Chinese
    shapefile = gpd.read_file('/Users/zheng/Desktop/时空数据挖掘/Data/tpu_data_geo/%s' % f, encoding="GBK")
    tupid = f.split('_')[1].split('.')[0]
    shapefile['tpuid'] = tupid
    ## set crs
    shapefile.crs = 'EPSG:4326'
    gfs.append(shapefile)
gfs = gpd.GeoDataFrame(pd.concat(gfs))
## select two columns only
sel_cols = ['tpuid', 'geometry']
gfs_sel = gfs[sel_cols]
```

```python
# 分类表读取
Group_df = pd.read_csv('/Users/zheng/Desktop/时空数据挖掘/Group_data.csv')
```

```python
merged_df = pd.merge(df, Group_df[['device_id', 'class', 'cluster']], on='device_id', how='inner')

# 保存 records 作为csv
# merged_df.to_csv('records.csv')
```

### 加入Geometry 与 TPU表 关联

```python
geometry = gpd.points_from_xy(merged_df['longitude'], merged_df['latitude'])
merged_gdf = gpd.GeoDataFrame(merged_df, geometry=geometry)
merged_gdf.crs = "EPSG:4326"
merged_gdf
```

```python
gfs_sel
```

```python
merged = gpd.sjoin(gfs_sel, merged_gdf, how='left', op='intersects')
merged['date'] = pd.to_datetime(merged["timestamp"], unit='s') + pd.Timedelta(days=8/24)
merged
```

```python
df_new = merged[['tpuid', 'geometry', 'device_id','date', 'class', 'cluster','latitude','longitude']]
df_new.crs = "EPSG:4326"
df_new
```
