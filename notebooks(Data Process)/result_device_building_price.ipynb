```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
import datetime
import matplotlib.pyplot as plt
import contextily as cx
```

## 读取 building_info #building_info_df

```python
path1 = 'process data/dataset1/'
os.listdir(path1)
```

```python
building_info_df = pd.read_csv(path1 + 'building_info.csv')
building_info_df
```

## 读取有住址的device #device_home_df

```python
path2 = 'process data/dataset2/'
os.listdir(path2)
```

```python
device_home_df = pd.read_csv(path2 + 'device_building_home.csv')
device_home_df
```

## 读取推测住址的device #device_pred_home_df

```python
path3 = 'process data/dataset3/'
os.listdir(path3)
```

```python
device_pred_home_df = pd.read_csv(path3 + 'device_building_location.csv')
device_pred_home_df
```

## 和并 device_home_df 和device_pred_home_df #device_df

```python
device_df = pd.concat([device_pred_home_df, device_home_df]).reset_index(drop=True)
device_df
```

```python
device_df[device_df['device_id'].duplicated()]
```

## 关联 device_df 和 building_info_df #device_price_df

```python
device_price_df = device_df.merge(building_info_df, on='BUILDINGST', how='inner')
device_price_df
```

```python
## 保存数据：device_price_result
device_price_df.to_csv("process data/device_price_result.csv", index=False)
```
